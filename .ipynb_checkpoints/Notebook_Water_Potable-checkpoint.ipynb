1. ph: pH of 1. water (0 to 14).
2. Hardness: Capacity of water to precipitate soap in mg/L.
3. Solids: Total dissolved solids in ppm.
4. Chloramines: Amount of Chloramines in ppm.
5. Sulfate: Amount of Sulfates dissolved in mg/L.
6. Conductivity: Electrical conductivity of water in μS/cm.
7. Organic_carbon: Amount of organic carbon in ppm.
8. Trihalomethanes: Amount of Trihalomethanes in μg/L.
9. Turbidity: Measure of light emiting property of water in NTU.
10. Potability: Indicates if water is safe for human consumption. Potable 1 and Not potable 0

Tomar distitnas opciones:
    1.- Eliminar Filas Con datos en nulos, pero, ¿Con cuantas filas quedare?, ¿Afectara mucho el modelo?
    2.- Poner el promedio de la columna en datos NaN, ¿Ruido en Diagrama de dispersion?, ¿Que regresion usar?, ¿Modelo quedara coherente?

### Mapa calor eliminando los NaN

```python
# guardar modelo eliminando filas NaN - Corroborar que no queden datos NaN
df_dropNaN = df.dropna()
df_dropNaN.isnull().sum()
```

```python
import numpy as np
import seaborn as snb
import matplotlib
import matplotlib.pyplot as plt

snb.set(font_scale=0.8)

df_dropNaN_T = df_dropNaN.values.T
_df_dropNaN = np.corrcoef(df_dropNaN_T)
 
_df_dropNaN = np.corrcoef(_df_dropNaN)
grafico = snb.heatmap(_df_dropNaN, annot=True)
plt.show()
```

### Mapa calor dejando el promedio en valores NaN

```python
# guardar modelo con el promedio - Corroborar que no queden datos NaN
df_meanNaN = df.fillna(df.mean(axis=0))
df_meanNaN.isnull().sum()
```

```python
df_meanNaN_T = df_meanNaN.values.T
_df_meanNaN = np.corrcoef(df_meanNaN_T)
 
_df_meanNaN = np.corrcoef(_df_meanNaN)
grafico = snb.heatmap(_df_meanNaN, annot=True)
plt.show()
```

### Mapa calor dejando 0 en valores NaN

```python
df_zeroNaN = df.fillna(0)
df_zeroNaN.isnull().sum()
```

```python
df_zeroNaN_T = df_zeroNaN.values.T
_df_zeroNaN = np.corrcoef(df_zeroNaN_T)
 
_df_zeroNaN = np.corrcoef(_df_zeroNaN)
grafico = snb.heatmap(_df_zeroNaN, annot=True)
plt.show()
```
